```python
import scipy
import scipy.io

# Load the MATLAB file
mat_data = scipy.io.loadmat('subject3_to_6_data2.mat')
```

```python
import numpy as np

data = mat_data['table']
print(data.shape) # shape of the table
print(data[2][0].shape) # all gestures from one subject
print(data[2][0][0].shape) # 7 gestures from one try
print(data[2][0][0][0].shape) # 1 hand gesture
print(data[2][0][0][0][0].shape) # 1 sensor
data = np.array(data)
```

```python
number_of_subject = (data.shape[0])
number_of_action_block = (data[2][0].shape)[0]
number_of_gestures_per_block = (data[2][0].shape)[1]
number_of_gestures_per_subject_ = (data[2][0].shape)[0]
number_of_sensors = (data[2][0][0][0][0].shape)[0]



print(number_of_subject)
print(number_of_action_block)
print(number_of_gestures_per_block)
print(number_of_sensors)
```

```python
print(data[2][0][2][0][:,8])
```

```python
# for each subject
for subject in range(2, number_of_subject):

  # for each action block
  for j in range(number_of_action_block):

    for gesture in range(number_of_gestures_per_block):

      for sensor in range(number_of_sensors):

        print(data[subject][0][j][gesture][:,sensor])
```

```python
import matplotlib.pyplot as plt
from scipy import signal

# Example EMG data (replace this with your actual data)
emg_signal_example = data[2][0][2][2][:,3]
sampling_rate = 5120

# Create a time vector based on the sampling rate and data length
time = np.arange(len(emg_signal_example)) / sampling_rate  # Time in seconds

# Plotting the original EMG signal
plt.figure(figsize=(10, 5))
plt.plot(time, emg_signal_example, color='b', linewidth=1)



# Adding labels and title
plt.xlabel("Time (s)")
plt.ylabel("EMG Signal (mV)")
plt.title("EMG Signal Over Time")
plt.grid(True)


# resample the signal
target_length = 15375
resampled_signal1 = signal.resample(emg_signal_example, target_length)
print(len(resampled_signal1))
resample_time = np.arange(target_length) / sampling_rate
print(resample_time)


# time_resampled = np.linspace(0, time[-1], target_length)
plt.plot(resample_time, resampled_signal1, color='r', linewidth=1)

plt.show()


```
